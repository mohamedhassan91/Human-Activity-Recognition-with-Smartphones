##### splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score,accuracy_score, classification_report
```

#### Training

```python
l_svm = LinearSVC(random_state=0)
svc = SVC(random_state=0)
lr = LogisticRegression(random_state=0)
knn = KNeighborsClassifier()
Rf = RandomForestClassifier(random_state=0)


# training
l_svm.fit(X_train, y_train)
svc.fit(X_train, y_train)
lr.fit(X_train, y_train)
knn.fit(X_train, y_train)
Rf.fit(X_train, y_train)
```

```python
for clf, label in zip([l_svm, svc, knn, lr,Rf], ['LinearSVC', 'SVC', 'KNeighborsClassifier', 'LogisticRegression','RandomForestClassifier']):
    print("{} report:".format(label))
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred))
    print("\n---------------------\n")
    
```

#### grid search to find best model

```python
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]

svc_grid = GridSearchCV(SVC(), tuned_parameters, cv=3, scoring='f1_micro', verbose=1, n_jobs=-1)

svc_grid.fit(X_train, y_train)
```
